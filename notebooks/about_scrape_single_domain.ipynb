```python
import requests
from bs4 import BeautifulSoup

test_url = "https://manifestationmatters.com/category/law-of-attraction/"
res = requests.get(test_url).text

soup = BeautifulSoup(res, "html.parser")
print(soup.prettify)
```

```python
tags = set([tag.name for tag in soup.find_all()])
print(tags)
```

```python
main = soup.find("main")
elements = main.find_all(['h1', 'h2', 'h3', 'p'])
content = []
for e in elements:
    content.append({
        "type": e.name,
        "text": e.get_text(strip=True)
    })

for c in content:
    print(c)
```
