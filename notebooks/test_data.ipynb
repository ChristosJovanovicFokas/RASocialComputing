```python
product_url = 'https://www.melissaoatman.com/store'
about_url = 'https://www.melissaoatman.com/'
```

```python
import sys
import os
from bs4 import BeautifulSoup

# Add project root to path
sys.path.append(os.path.abspath(".."))

from core.fetcher import Fetcher

fetcher = Fetcher()
#Fetch and parse single website
product_html = fetcher.fetch(product_url)
about_html = fetcher.fetch(about_url)
```

```python
from core.storage import *
```

```python
product_page_id = make_filename(product_url)
about_page_id = make_filename(about_url)
```

```python
product_main_path = save(product_html, "main.html", subfolder=f"products/{product_page_id}")
about_main_path = save(about_html, "main.html", subfolder=f"about/{about_page_id}")
```

```python
from core.parser import Parser

parser = Parser(product_html)

#parse product items and css for product
soup = BeautifulSoup(product_html, 'html.parser')
ul = soup.find(
    "ul",
    {
        "class": "S4WbK_"
    }
)
product_list = ul.find_all('li') if ul else []

product_css = parser.parse_css()
```

```python
print(ul)
```

```python
print(product_list)
```

```python
product_css_path = save(
    product_css,
    f"{product_page_id}.css",
    subfolder="css"
)
```

```python
#Parse about section and css
about_parser = Parser(about_html)
about_soup = BeautifulSoup(about_html, 'html.parser')

section = about_soup.find(
    "main",
    {
        "id": "PAGES_CONTAINER"
    }
)

about_css = about_parser.parse_css()
```

```python
print(section)
```

```python
#Save about section 

if section:
    about_section_path = save(
        str(section),
        f"{about_page_id}_section.html",
        subfolder=f"about/{about_page_id}"
    )
```

```python
#Save about css
about_css_path = save(
    str(about_css),
    f"{about_page_id}.css",
    subfolder="css"
)
```

```python
#Save individual product
for i, product in enumerate(product_list):
    product_id = make_filename(f"{product_url}_{i}")
    
    product_path = save(
        str(product),
        f"{product_id}.html",
        subfolder=f"products/{product_page_id}/items"
    )

    record = {
        "product_id": product_id,

        "product_html": product_path,
        "product_page_html": product_main_path,
        "product_css": product_css_path,

        "about_html": about_main_path,
        "about_section": about_section_path,
        "about_css": about_css_path,

        "source_product_url": product_url,
        "source_about_url": about_url
    }

    append_jsonl(record)
```
